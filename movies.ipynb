```python
# Import dependencies
import pandas as pd
```

```python
# Store csv file paths in variables
movies = "Resources/movies.csv"
```

```python
# Import data from csv files
movies_df = pd.read_csv(movies)
movies_df
```

```python
clean_movies_df = movies_df.drop(columns=['released','runtime','score','votes','writer'])
clean_movies_df.count()
```

```python
clean_movies_df = clean_movies_df.dropna(how='all', axis=0)
clean_movies_df.count()
```

```python
clean_movies_df= clean_movies_df[clean_movies_df['budget'] != 0]
clean_movies_df.count()
```

```python
clean_movies_df.to_csv("clean_movies.csv")
```
